And for other two columns last_review and no_of_reviews per month , both the columns are null when no_of reviews is zero . So , I replaced missing reviews per month with 0 and last_review with mode value of the column as anyways date wont make much sense when there is no review at all

```python
# Group columns for encoded neighborhood groups and room types
neighborhood_cols = [
    'neighbourhood_group_Bronx', 'neighbourhood_group_Brooklyn',
    'neighbourhood_group_Manhattan', 'neighbourhood_group_Queens',
    'neighbourhood_group_Staten Island'
]
room_type_cols = [
    'room_type_Entire home/apt', 'room_type_Private room', 'room_type_Shared room'
]
```

```python
# Calculate the frequency of room types for each neighborhood
room_type_freq = pd.DataFrame({
    neighborhood: df[room_type_cols].mul(df[neighborhood], axis=0).sum()
    for neighborhood in neighborhood_cols
})

room_type_freq = room_type_freq.T
room_type_freq.columns = ['Entire home/apt', 'Private room', 'Shared room']

# Plot bar graphs for each neighborhood group
fig, axes = plt.subplots(1, 5, figsize=(20, 5), sharey=True)

for ax, neighborhood in zip(axes, neighborhood_cols):
    room_type_freq.loc[neighborhood].plot(
        kind='bar', ax=ax, color=['skyblue', 'lightgreen', 'salmon']
    )
    ax.set_title(neighborhood.replace('neighbourhood_group_', ''))
    ax.set_xlabel('Room Type')
    ax.set_ylabel('Frequency')
    ax.set_xticklabels(['Entire home/apt', 'Private room', 'Shared room'], rotation=45)

plt.tight_layout()
plt.show()
```

# Outlier detection and handling

```python
plt.figure(figsize = (16,5))
sns.distplot(df['price'])
```

```python
df['price'].skew()
```

Since the data is skew , we cant use Z score . Therefore I'm using IQR

```python
df['price'].describe()
```

```python
sns.boxplot(df['price'])
```

```python
# Finding the IQR
percentile25 = df['price'].quantile(0.25)
percentile75 = df['price'].quantile(0.75)
```

```python
percentile25
```

```python
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
df['price'].max()
```

```python
df['price'].min()
```

```python
df[df['price'] > upper_limit]
```

```python
new_df = df[df['price'] < upper_limit]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

features = df[['availability_365', 'minimum_nights', 'number_of_reviews', 
               'neighbourhood_group_Bronx', 'neighbourhood_group_Brooklyn', 
               'neighbourhood_group_Manhattan', 'neighbourhood_group_Queens', 
               'neighbourhood_group_Staten Island', 'room_type_Entire home/apt', 
               'room_type_Private room', 'room_type_Shared room']]
target = df['price']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train Random Forest Regressor
model = RandomForestRegressor(random_state=42, n_estimators=100)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluate the model
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R Score: {r2}")

# Feature importance visualization
importances = model.feature_importances_
plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features.columns, palette="viridis")
plt.title('Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
